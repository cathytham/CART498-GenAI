```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
#import library

from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "Lykon/dreamshaper-8",
    torch_dtype=torch.float16,
    variant="fp16",
    ).to(device)
```

```python
prompt = "A golden Labrador running joyfully through a field of beautiful flowers that are melting into a puddle of colorful liquid, with an early morning sunlight creating a warm feeling. In the surreal style of Max Ernst."
image =pipe(prompt).images[0]

image.save("surreal_collage_1.png")
```

```python
prompt="A golden Labrador running through a field of flowers at sunrise where the flowers are melting and dripping to the ground into different shapes. Similar to a strange dream with unusual textures in the surreal style of Max Ernst."
image = pipe(prompt).images[0]

image.save("surreal_collage_2.png")
```
